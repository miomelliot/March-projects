```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
#       
n = 100
steps = np.random.normal(loc=0, scale=1, size=n)
random_walk = steps.cumsum()

plt.figure(figsize=(10, 4))
plt.plot(random_walk)
plt.title(" ")
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.show()
```

```python
#        DataFrame
steps1 = np.random.normal(loc=0, scale=1, size=n).cumsum()
steps2 = np.random.normal(loc=0, scale=1, size=n).cumsum()

df = pd.DataFrame({
    'Walk A': steps1,
    'Walk B': steps2
})

plt.figure(figsize=(10, 4))
plt.plot(df['Walk A'], label=' A', color='orange')
plt.plot(df['Walk B'], label=' B', color='green')
plt.title("  ")
plt.xlabel("")
plt.ylabel("")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#      2025
date_range = pd.date_range(start='2025-01-01', end='2025-02-28')
steps3 = np.random.normal(loc=0, scale=1, size=len(date_range)).cumsum()
df_dates = pd.DataFrame({'Value': steps3}, index=date_range)

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(df_dates, label=' ', color='purple')

ax.set_title("    2025")
ax.set_xlabel("")
ax.set_ylabel("")
ax.legend()

from matplotlib.ticker import MultipleLocator
#    
ax.xaxis.set_major_locator(MultipleLocator(10))  #    10 
ax.xaxis.set_minor_locator(MultipleLocator(1))   #    
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
