```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
# Сгенерируем одно случайное блуждание и визуализируем его
n = 100
steps = np.random.normal(loc=0, scale=1, size=n)
random_walk = steps.cumsum()

plt.figure(figsize=(10, 4))
plt.plot(random_walk)
plt.title("Случайное блуждание")
plt.xlabel("Итерации")
plt.ylabel("Значение")
plt.grid(True)
plt.show()
```

```python
# Сгенерируем два случайных блуждания и превратим в DataFrame
steps1 = np.random.normal(loc=0, scale=1, size=n).cumsum()
steps2 = np.random.normal(loc=0, scale=1, size=n).cumsum()

df = pd.DataFrame({
    'Walk A': steps1,
    'Walk B': steps2
})

plt.figure(figsize=(10, 4))
plt.plot(df['Walk A'], label='Маршрут A', color='orange')
plt.plot(df['Walk B'], label='Маршрут B', color='green')
plt.title("Два случайных блуждания")
plt.xlabel("Итерации")
plt.ylabel("Значение")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Случайное блуждание с датами январь–февраль 2025
date_range = pd.date_range(start='2025-01-01', end='2025-02-28')
steps3 = np.random.normal(loc=0, scale=1, size=len(date_range)).cumsum()
df_dates = pd.DataFrame({'Value': steps3}, index=date_range)

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(df_dates, label='Случайное блуждание', color='purple')

ax.set_title("Случайное блуждание за Январь–Февраль 2025")
ax.set_xlabel("Дата")
ax.set_ylabel("Значение")
ax.legend()

from matplotlib.ticker import MultipleLocator
# Основные и промежуточные деления
ax.xaxis.set_major_locator(MultipleLocator(10))  # Основные – каждые 10 дней
ax.xaxis.set_minor_locator(MultipleLocator(1))   # Промежуточные – каждый день
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
