```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Загрузка данных
titanic_df = pd.read_csv('train.csv')
```

```python
# Предпросмотр данных
titanic_df.head()
```

```python
# Информация о датасете
titanic_df.info()
```

```python
# Общее количество записей и столбцов
print(f'Всего записей: {titanic_df.shape[0]}, Всего столбцов: {titanic_df.shape[1]}')
```

```python
# Основные статистические параметры
titanic_df.describe(include='all')
```

**Выводы:**
- Всего 891 запись и 12 столбцов.
- Присутствуют пропуски в Age, Cabin, Embarked.
- Средний возраст пассажиров ~30 лет, цена билета варьируется значительно.
- Большинство пассажиров садились в порту "S".

```python
sns.catplot(x='Pclass', hue='Survived', kind='count', data=titanic_df, palette='Set2')
plt.title('Выживаемость по классу билета')
```

```python
sns.catplot(x='Sex', hue='Survived', kind='count', data=titanic_df, palette='Set1')
plt.title('Выживаемость по полу')
```

```python
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=titanic_df, palette='pastel')
plt.title('Вероятность выжить в зависимости от класса и пола')
```

```python
sns.catplot(x='Embarked', hue='Survived', kind='count', data=titanic_df, palette='coolwarm')
plt.title('Выживаемость по порту посадки')
```

```python
sns.catplot(x='Embarked', y='Survived', hue='Pclass', kind='bar', data=titanic_df)
plt.title('Выживаемость по классу и порту посадки')

```

```python
sns.barplot(x='Parch', y='Survived', hue='Parch', data=titanic_df, palette='Set2', legend=False)
plt.title('Средняя выживаемость по количеству родителей и детей')
```

```python
sns.catplot(x='SibSp', hue='Survived', kind='count', data=titanic_df)
plt.title('Выживаемость по количеству братьев/сестёр/супругов')
```

```python
sns.displot(x='Age', hue='Survived', kind='kde', fill=True, data=titanic_df)
plt.title('Распределение выживаемости по возрасту')
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(titanic_df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Корреляционная матрица признаков')
```

### 📊 Выводы по корреляционной матрице признаков

- **`Pclass` и `Fare`** имеют **сильную отрицательную корреляцию** (`-0.55`)  
  → чем выше класс (т.е. ближе к 1), тем дороже билет. Логично: 1-й класс → высокие цены.

- **`Survived` и `Fare`** имеют **умеренную положительную корреляцию** (`0.26`)  
  → пассажиры с более дорогими билетами чаще выживали.

- **`Survived` и `Pclass`** — **умеренная отрицательная корреляция** (`-0.34`)  
  → пассажиры 1-го класса выживали чаще, а 3-го — реже.

- **`SibSp` и `Parch`** — **умеренно положительная корреляция** (`0.41`)  
  → если у пассажира есть братья/сёстры, у него часто есть и родители/дети на борту.

- **`Age` слабо отрицательно коррелирует с `Pclass` и `SibSp`**  
  → молодые чаще ехали в 3-м классе и с семьёй.

- **`PassengerId`** практически ни с чем не коррелирует — это ожидаемо, ведь это просто номер строки.

---

### 🧠 Общий вывод:

- Выживаемость сильнее всего зависит от **класса билета** и **стоимости билета**, то есть от социального статуса.
- Также (в меньшей степени) выживаемость связана с **наличием семьи на борту**.
- Возраст влияет, но это слабее выражено в линейной корреляции — нужно смотреть на распределение.
